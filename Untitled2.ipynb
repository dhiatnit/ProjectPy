# tkinter

```python
import tkinter as tk
root = tk.Tk()
root.title("My first  window")


welcome = tk.Label(root,text="Welcome to tkinter")
welcome.pack()

def on_button():
    welcome.config(text="Changed")

btn = tk.Button(root,text="Click me !",command=on_button)
btn.pack()
root.mainloop()
```
